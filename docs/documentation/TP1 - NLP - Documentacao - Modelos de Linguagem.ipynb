# Análises

## Tamanhos de corpus

Para efeitos de variação do tamanho do corpus, foram utilizadas diferentes porcentagens:

20%, 40%, 60%, 80% e 100% do tamanho total do corpus contido no text8.

Os tamanhos das janelas de contexto foram mantidas como 8 para todos os modelos.

```python
import matplotlib.pyplot as plt
import numpy as np

cbowModels = []
cbowCorpusMatches = []
cbowCorpusDistanceMean = []

# 20% -> cbow5
cbowModels.append('cbow5')
cbowCorpusMatches.append(cbowMatches[4])
cbowCorpusDistanceMean.append(cbowDistancesMean[4])

# 40% -> cbow6
cbowModels.append('cbow6')
cbowCorpusMatches.append(cbowMatches[5])
cbowCorpusDistanceMean.append(cbowDistancesMean[5])

# 60% -> cbow7
cbowModels.append('cbow7')
cbowCorpusMatches.append(cbowMatches[6])
cbowCorpusDistanceMean.append(cbowDistancesMean[6])

# 80% -> cbow8
cbowModels.append('cbow8')
cbowCorpusMatches.append(cbowMatches[7])
cbowCorpusDistanceMean.append(cbowDistancesMean[7])

# 100% -> cbow3
cbowModels.append('cbow3')
cbowCorpusMatches.append(cbowMatches[2])
cbowCorpusDistanceMean.append(cbowDistancesMean[2])
```

```python
skipModels = []
skipCorpusMatches = []
skipCorpusDistanceMean = []

# 20% -> skip5
skipModels.append('skip5')
skipCorpusMatches.append(skipMatches[4])
skipCorpusDistanceMean.append(skipDistancesMean[4])

# 40% -> skip6
skipModels.append('skip6')
skipCorpusMatches.append(skipMatches[5])
skipCorpusDistanceMean.append(skipDistancesMean[5])

# 60% -> skip7
skipModels.append('skip7')
skipCorpusMatches.append(skipMatches[6])
skipCorpusDistanceMean.append(skipDistancesMean[6])

# 80% -> skip8
skipModels.append('skip8')
skipCorpusMatches.append(skipMatches[7])
skipCorpusDistanceMean.append(skipDistancesMean[7])

# 100% -> skip3
skipModels.append('skip3')
skipCorpusMatches.append(skipMatches[2])
skipCorpusDistanceMean.append(skipDistancesMean[2])
```

```python
plt.bar(cbowModels, cbowCorpusMatches)
plt.ylim(0, max(cbowCorpusMatches)*1.1)
plt.text(-.2, cbowCorpusMatches[0]+200, str(cbowCorpusMatches[0]), fontsize=12)
plt.text(0.75, cbowCorpusMatches[1]+200, str(cbowCorpusMatches[1]), fontsize=12)
plt.text(1.7, cbowCorpusMatches[2]+200, str(cbowCorpusMatches[2]), fontsize=12)
plt.text(2.7, cbowCorpusMatches[3]+200, str(cbowCorpusMatches[3]), fontsize=12)
plt.text(3.7, cbowCorpusMatches[4]+200, str(cbowCorpusMatches[4]), fontsize=12)
plt.title('Total Matches for each CBOW model with different corpus\' sizes')
plt.xlabel('Model')
plt.ylabel('Matches')
plt.show()
```

```python
plt.bar(skipModels, skipCorpusMatches)
plt.ylim(0, max(skipCorpusMatches)*1.1)
plt.text(-.2, skipCorpusMatches[0]+200, str(skipCorpusMatches[0]), fontsize=12)
plt.text(0.75, skipCorpusMatches[1]+200, str(skipCorpusMatches[1]), fontsize=12)
plt.text(1.7, skipCorpusMatches[2]+200, str(skipCorpusMatches[2]), fontsize=12)
plt.text(2.7, skipCorpusMatches[3]+200, str(skipCorpusMatches[3]), fontsize=12)
plt.text(3.7, skipCorpusMatches[4]+200, str(skipCorpusMatches[4]), fontsize=12)
plt.title('Total Matches for each Skip-gram model with different corpus\' sizes')
plt.xlabel('Model')
plt.ylabel('Matches')
plt.show()
```

```python
proportion = list(map(lambda x, y: x/y, cbowCorpusMatches, skipCorpusMatches))

np.mean(proportion)
```

### Análise dos resultados

Esse resultado anterior (proportion) indica que, em média, o modelo CBOW consegue realizar 10.47% matches a mais que o modelo Skip-gram, demonstrando que ele consegue gerar mais saídas com os mesmos tamanhos de corpus.

```python
plt.plot(cbowModels, cbowCorpusDistanceMean, 'g--o')
for model, dist in zip(cbowModels, cbowCorpusDistanceMean):
    plt.annotate(str(round(dist, 6)),xy=(model, dist+0.0005))
plt.ylim(min(cbowCorpusDistanceMean) * 0.9, max(cbowCorpusDistanceMean) * 1.05)
plt.title('Distances Mean for each CBOW model with different corpus\' sizes')
plt.xlabel('Model')
plt.ylabel('Distances Mean')
plt.show()
```

```python
plt.plot(skipModels, skipCorpusDistanceMean, 'g--o')
for model, dist in zip(skipModels, skipCorpusDistanceMean):
    plt.annotate(str(round(dist, 6)),xy=(model, dist+0.0005))
plt.ylim(min(skipCorpusDistanceMean) * 0.9, max(skipCorpusDistanceMean) * 1.05)
plt.title('Distances Mean for each Skip-gram model with different corpus\' sizes')
plt.xlabel('Model')
plt.ylabel('Distances Mean')
plt.show()
```

```python
proportion = list(map(lambda x, y: x/y, cbowCorpusDistanceMean, skipCorpusDistanceMean))

np.mean(proportion)
```

### Análise dos resultados

O resultado acima (proportion), demonstra que, em média, as distâncias médias do modelo CBOW é 15.11% maior que as distâncias médias no modelo Skip-gram.

Podemos concluir também que, conforme o tamanho do corpus aumenta, o desempenho do algoritmo também melhora, demonstrando que o tamanho do corpus e as distâncias médias são inversamente proporcionais. Isso era de se esperar, pois como em qualquer outro modelo de aprendizagem de máquina, quanto mais dados você possui para treinar o seu modelo, maior a acurácia das previsões que ele realiza.

### Análise dos Resultados

O resultado acima (proportion) demonstra que, em média, o CBOW possui distâncias médias 22.85% maiores que o Skip-gram para os mesmos tamanhos de janela de contexto.

Podemos concluir também que para o CBOW, tamanhos da janela de contexto entre 4 e 10 apresentam um bom desempenho e acurácia. Isso pode estar relacionado ao fato de que como o CBOW utiliza do contexto para prever a saída correta, quanto maior o contexto, maior a eficácia dessa previsão.

Já para o Skip-gram, tamanhos de janela de contexto entre 2 e 4 apresentam um bom desempenho e acurácia. Isso pode estar relacionado ao fato de que como o Skip-gram utiliza da palavra para prever o contexto da saída correta, quanto menor o contexto, mais fácil de prevê-lo e isso aumenta a eficácia do modelo.

## Comparação entre os modelos CBOW e Skip-gram

Para comparação desses modelos, serão utilizados os 8 modelos citados acima.
Todos os pares de mesmo índice (cbowi, skipi) possuem exatamente os mesmos tamanhos de corpus e
janelas de contexto. O que os diferem é somente o modelo utilizado.

```python
plt.plot(list(range(1,9)), cbowMatches, 'r--o', label='CBOW')
plt.plot(list(range(1,9)), skipMatches, 'go-', label='Skip-gram')
plt.legend(loc=3)
plt.title('Comparison between CBOW and Skip-gram using number of matches')
plt.xlabel('Model index')
plt.ylabel('Matches')
plt.show()
```

```python
proportion = list(map(lambda x, y: x/y, cbowMatches, skipMatches))

np.mean(proportion)
```

### Análise dos resultados

Conforme podemos observar acima, o modelo CBOW possui, em média, 8.79% mais matches que o modelo Skip-gram. Portanto, ele consegue dar mais saídas (matches), se mostrando um modelo mais abrangente.

```python
plt.plot(list(range(1,9)), cbowDistancesMean, 'r--o', label='CBOW')
plt.plot(list(range(1,9)), skipDistancesMean, 'go-', label='Skip-gram')
plt.legend(loc=2)
plt.show()
```

```python
proportion = list(map(lambda x, y: x/y, cbowDistancesMean, skipDistancesMean))

np.mean(proportion)
```

### Análise dos resultados

Conforme podemos observar acima, o modelo CBOW possui, em média, distâncias médias 18.86% maior que o modelo Skip-gram. Portanto, ele é menor assertivo nas suas respostas, se mostrando um modelo menos eficiente em relação a retornar a palavra "target" com uma maior similaridade de cosseno (medido como distância de cosseno no algoritmo).

# Conclusões

O modelo CBOW se mostrou um modelo mais abrangente, pois para as mesmas configurações de parâmetros, ele consegue retornar mais respostas (matches). Porém, ao mesmo tempo, ele possui uma menor acurácia, pois para as mesmas configurações de parâmetros, as suas distâncias médias são maiores que as do modelo Skip-gram.

Ao longo da execução dos testes, também percebi que o modelo Skip-gram demora mais para ser treinado que o modelo CBOW, podendo isso ser uma justificativa plauśivel pela maior acurácia, pois seu 'learning rate' (ou alpha, no caso dos parâmetros) é menor que o CBOW.

Portanto, caso seja necessário um modelo de treino mais rápido e mais abrangente, deverá ser utilizado o CBOW. Caso seja necessário um modelo de treino que possui uma maior acurácia nas suas respostas, deverá ser utilizado o Skip-gram.
